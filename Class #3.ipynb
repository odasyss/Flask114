```python
def fact(n):
    if n==0:
        return 1
    return n*fact(n-1)
```

```python
fact(5)
```

```python
from functools import lru_cache

@lru_cache
def fib(n):
    if n<2:
        return n
    return fib(n-1) + fib(n-2)
```

```python
fib(7)
```
